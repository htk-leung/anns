```python
#%%
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

def loaddata(filename):
    arr = np.loadtxt(filename, delimiter=',', skiprows=1)
    return {
        'd':                arr[:, 0],
        'efc':              arr[:, 1],
        'm':                arr[:, 2],
        'efs':              arr[:, 3],
        'k':                arr[:, 4],
        'recall':           arr[:, 5],
        'distcountconstr':  arr[:, 6],
        'distcountsearch':  arr[:, 7]
    }

xparam = "recall"
yparam = "distcountsearch"

hnswsiftc = loaddata('../../w05/results/data/20251002-hnsw-dist-cache-sift.txt')

#s100 c100
flatnavsift = loaddata('../../w05/results/data/20251002-flatnav-sift.txt')
flatnavsiftRc = loaddata('data/20251004-flatnav-sift-cRand-s100c100.txt')
flatnavsiftR = loaddata('data/20251007-flatnav-sift-scRand-s100c100.txt') 
flatnavsiftRs = loaddata('data/20251007-flatnav-sift-sRand-s100c100.txt')

#s100 c50
flatnavsift10050 = loaddata('../../w05/results/data/20251002-flatnav-sift-s100-c50.txt') 
flatnavsift10050Rc = loaddata('data/20251004-flatnav-sift-cRand-s100c50.txt')
flatnavsift10050R = loaddata('data/20251007-flatnav-sift-scRand-s100c50.txt')
flatnavsift10050Rs = loaddata('data/20251007-flatnav-sift-sRand-s100c50.txt')

#s100 c25
flatnavsift10025 = loaddata('data/20251004-flatnav-sift-s100c25.txt')
flatnavsift10025Rc = loaddata('data/20251004-flatnav-sift-cRand-s100c25.txt') 
flatnavsift10025R = loaddata('data/20251007-flatnav-sift-scRand-s100c25.txt')
flatnavsift10025Rs = loaddata('data/20251007-flatnav-sift-sRand-s100c25.txt')

#s100 c10
flatnavsift10010 = loaddata('data/20251004-flatnav-sift-s100c10.txt')
flatnavsift10010Rc = loaddata('data/20251004-flatnav-sift-cRand-s100c10.txt') 
flatnavsift10010R = loaddata('data/20251007-flatnav-sift-scRand-s100c10.txt')
flatnavsift10010Rs = loaddata('data/20251007-flatnav-sift-sRand-s100c10.txt')

#s100 c200
flatnavsift100200 = loaddata('../../w05/results/data/20251002-flatnav-sift-s100-c200.txt')
flatnavsift100200Rc = loaddata('data/20251004-flatnav-sift-cRand-s100c200.txt')



#s50 c100
flatnavsift50100 = loaddata('data/20251007-flatnav-sift-s50c100.txt') # missing
flatnavsift50100Rc = loaddata('data/20251007-flatnav-sift-cRand-s50c100.txt')
flatnavsift50100R = loaddata('data/20251007-flatnav-sift-scRand-s50c100.txt')
flatnavsift50100Rs = loaddata('data/20251007-flatnav-sift-sRand-s50c100.txt')

#s25 c100
flatnavsift25100 = loaddata('data/20251007-flatnav-sift-s25c100.txt') # missing
flatnavsift25100Rc = loaddata('data/20251007-flatnav-sift-cRand-s25c100.txt') 
flatnavsift25100R = loaddata('data/20251007-flatnav-sift-scRand-s25c100.txt')
flatnavsift25100Rs = loaddata('data/20251007-flatnav-sift-sRand-s25c100.txt')

#s10 c100
flatnavsift10100 = loaddata('data/20251007-flatnav-sift-s10c100.txt') # missing
flatnavsift10100Rc = loaddata('data/20251007-flatnav-sift-cRand-s10c100.txt') 
flatnavsift10100R = loaddata('data/20251007-flatnav-sift-scRand-s10c100.txt')
flatnavsift10100Rs = loaddata('data/20251007-flatnav-sift-sRand-s10c100.txt')
```

```python
def plotsamenum(regular, rands, randc, randsc, title, ins, inc):
    x1 = regular[xparam]
    y1 = regular[yparam]
    x2 = rands[xparam]
    y2 = rands[yparam]
    x3 = randc[xparam]
    y3 = randc[yparam]
    x4 = randsc[xparam]
    y4 = randsc[yparam]

    # create plot
    fig, ax = plt.subplots()

    # set y-axis to logarithmic scale
    ax.set_yscale('log') 

    # plot
    ax.scatter(x1, y1, marker='x', color='black', s=50, label=f's{ins}c{inc}stepsize', alpha=1)
    ax.scatter(x2, y2, marker='x', color='red', s=50, label=f's{ins}randc{inc}', alpha=0.7)
    ax.scatter(x3, y3, marker='x', color='blue', s=50, label=f's{ins}c{inc}rand', alpha=0.7)
    ax.scatter(x4, y4, marker='x', color='green', s=50, label=f's{ins}randc{inc}rand', alpha=0.7)

    # display
    ax.set_xlabel(xparam)
    ax.set_ylabel('search distance computation count')
    ax.set_title(title)
    ax.legend()
    ax.grid(True, alpha=0.3)

    #display graph
    plt.show()
```

```python
title = 'flatnav random sampling vs stepsize visit s100c100'
plotsamenum(flatnavsift, flatnavsiftRs, flatnavsiftRc, flatnavsiftR, title, 100, 100)
```
